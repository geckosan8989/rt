```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'/content/IMDB Dataset - IMDB Dataset.csv')
df.head(3)
```

```python
#replace
df['sentiment'].replace({'positive': 1, 'negative': 0},inplace=True)
```

```python
x=np.array(df['review'].values)
y=np.array(df['sentiment'].values)
```

```python
#fitering sentences
x_filtered=[]

for review in x:
  review=review.lower()

  for i in review:
    punc=''' !()-[]{};:'"\,<>./?@#$%^&*_~ '''
    if i in punc:
      review=review.replace(i,'')


  x_filtered.append(review)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer,one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocabulary_size = 5000
one_hot_encoded= [one_hot(review, vocabulary_size) for review in x_filtered]
print(one_hot_encoded)
```

```python
max_length=500
x_padded=pad_sequences(one_hot_encoded, maxlen=max_length,padding='post')
```

```python
#Train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_padded,y,test_size=0.3,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
embedded_vector_size=35

model=Sequential()
model.add(Embedding(vocabulary_size,embedded_vector_size,input_length=max_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
```

```python
print(model.summary())
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=1)
```

```python
score=model.evaluate(x_test,y_test)
print('Test Loss:',score[0])
print('Test Accuracy:',score[1])
```
